Next step, we gonna calculate percentage of missing data we have in datasets. For this task, we split datasets in one (as mentioned in previous stage):

```python
train_data_n_size = train_data.shape[0]
concat_sets_data = pd.concat((train_data, test_data)).reset_index(drop=True)
ytrain = train_data.SalePrice.values
concat_sets_data.drop(['SalePrice'], axis=1, inplace=True)
print("concat_sets_data size is : {}".format(concat_sets_data.shape))
```

```python
def missing_ratio_df(all_data):
    all_data_na = (all_data.isnull().sum() / len(all_data)) * 100
    all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)
    missing_data_perc = pd.DataFrame({'Missing Ratio' :all_data_na})
    return missing_data_perc
```

```python
missing_data_perc = missing_ratio_df(concat_sets_data)
missing_data_perc
```

Now, we drop columns, which have *missing ratio* bigger than 75%:

```python
for row_name, row_data in missing_data_perc.iterrows():
    if row_data['Missing Ratio'] >= 75.0:
        concat_sets_data.drop(row_name, axis = 1, inplace = True)
```

```python
missing_data_perc = missing_ratio_df(concat_sets_data)
missing_data_perc
```

Let's impute remaining missing data with appropriate data. But, for this type of activity, we need to know context and meaning of the columns. After reading data-description.txt we can say that we replace:
 
* *'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType', 'MSSubClass'* with **None**, because this variables are categorical;

* *LotFrontage: Linear feet of street connected to property* we can replace it with **median** value based on neightborhood.

* *'GarageYrBlt', 'GarageArea', 'GarageCars', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath', 'BsmtHalfBath', 'MasVnrArea'* with **0**;

* *'MSZoning', 'Electrical', 'KitchenQual', 'Exterior1st', 'Exterior2nd', 'SaleType'* with **mode** and *'Functional*' with **Typ**;

```python
for col in ('FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'BsmtQual', 
            'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType', 'MSSubClass'):
    concat_sets_data[col] = concat_sets_data[col].fillna('None')
concat_sets_data.drop(['Utilities'], axis=1, inplace = True)

concat_sets_data["LotFrontage"] = concat_sets_data.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median()))
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF','TotalBsmtSF', 'BsmtFullBath',
            'BsmtHalfBath', 'MasVnrArea'):
    concat_sets_data[col] = concat_sets_data[col].fillna(0)

for col in ('MSZoning', 'Electrical', 'KitchenQual', 'Exterior1st', 'Exterior2nd', 'SaleType'):
    concat_sets_data[col] = concat_sets_data[col].fillna(concat_sets_data[col].mode()[0])

concat_sets_data["Functional"] = concat_sets_data["Functional"].fillna("Typ")
```

```python
missing_data_perc = missing_ratio_df(concat_sets_data)
missing_data_perc
```

# 3. Spliting data and Building a model

Splitting data to train and test set using the previous ratio:

```python
train_data_ = concat_sets_data[:train_data_n_size]
test_data_ = concat_sets_data[train_data_n_size:]
```

We can see that data mostly have simple linear and quadratic relation. Predictable variable is numerical. Also, there area a lot of highly correlated data. To cover all features and behaviours in data, we can stack results of several models. 

The basical approach of this problem is to use simple *multiple linear regression*. But one of the main problem is that some data have quadratic relation between other parameters. To solve that, we can add *polinomial regression* (*quadratic* in our case) algorithm.  Also, as we say before, there many parameters which having high correlation value, it can be the main factor of model overfitting. To avoid this we can use *multiple ridge regression* instead of *multiple linear regression* and "multiple kernel ridge regression" instead of "multiple polinomial regression". This algorithm is good for dealing with overfitting because it uses regularization technic called L2 Regularization. But, there are algorithm which combine two regularization technics (L1, L2 Regularization), but using linear regression as main algorithm. This algorithm called *Elastic Net regression*. We use it instead of *multiple ridge regression*.

The last regression model which we going to use is a *gradient boosting*. Gradient boosting is a technique for regression roblem, which constructs a prediction model in the form of an ensemble of weak prediction models, typically decision trees which become better through iterations. Gradient boosting is a greedy algorithm. It means that it can overfit a training dataset quickly. It can benefit from regularization methods that penalize various parts of the algorithm and generally improve the performance of the algorithm by reducing overfitting.

**About metrics**

There are 3 standart metrics for regression problems:
* Mean Squared Error (MSE)
* Root Mean Squared Error (RMSE)
* Mean Absolute Error (MAE)

This metrics widely used in regression problems. We gonna use RMSE, because RMSE has the benefit of penalizing large errors more that other 2 metrics. MSE and RMSE punish larger errors more than smaller errors, inflating or magnifying the mean error score. This is due to the square of the error value. The MAE does not give more or less weight to different types of errors and instead the scores increase linearly with increases in error.

Also, we use $R^2$ score. $R^2$ is a statistical measure of how close the data are to the fitted regression line. It is also known as the coefficient of determination, or the coefficient of multiple determination for multiple regression. In simple words, coefficient of determination meassures the performance of regression model. $R^2$ is in the range from $-2$ to $1$, where $-2$ stands for the worst result, $0$ for average result and $1$ for perfect performance.

The model is gonna be average result of 3 regression models:
* Kernel Ridge Regression
* ElasticNet
* Gradient Boosting

```python
# We use Robust Scaler because standardization of a dataset is a common
# requirement for many machine learning estimators.
enet_regr = make_pipeline(RobustScaler(), ElasticNet(alpha = 0.001, l1_ratio = 0.95))
krr_regr = KernelRidge(alpha = 0.65, kernel = 'polynomial', degree = 2, coef0=2.5)
gb_regr = GradientBoostingRegressor(n_estimators=3000, learning_rate = 0.05,
                                   max_depth = 4, max_features = 'sqrt',
                                   min_samples_leaf = 15, min_samples_split = 10)
```

```python
# Using np.log1p for make values smaller
ytrain = np.log1p(ytrain)
```

```python
# Class for models
class AverageModel(BaseEstimator, RegressorMixin):
    
    # Init model (and models for averaging)
    def __init__(self, models_list):
        super().__init__()
        
        self.models = models_list
        self.fitted_models = []
    
    # Fit model to the train data (X to Y)
    # We clone models if we want to reuse
    # model on new data
    def fit(self, X, Y):
        for model in self.models:
            self.fitted_models.append(clone(model).fit(X, Y))
            
        return self
    
    # Predict Y based on test X data (and average them)
    def predict(self, X):
        predictions = np.column_stack([model.predict(X) for model in self.fitted_models])
        
        return np.mean(predictions, axis=1)
```

```python
models_list = [enet_regr, krr_regr, gb_regr]
avg_model = AverageModel(models_list)

avg_model.fit(train_data_, ytrain)
preds = avg_model.predict(test_data_)
```
